```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
Books = []
for i in range(1,51):
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    r = requests.get(url).text
    soup = BeautifulSoup(r, "html.parser")
    ol = soup.find('ol')
    article = ol.findAll('article', class_= 'product_pod')
    for i in article:
      image = i.find('img')
      title = image.attrs['alt']
      stars = i.find('p')
      stars = stars['class'][1]
      price = i.find('p', class_ = 'price_color').text.replace('Â','')
      price = float((price)[1:])
      Books.append([title,price,stars])

Books_df = pd.DataFrame(Books,columns=['Title','Price','Stars Rating'])
Books_df
```
